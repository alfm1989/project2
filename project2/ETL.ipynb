```python
import pandas as pd
from sqlalchemy import create_engine
```

### Extract CSVs into DataFrames

```python
confirmados_file = "Confirmados.csv"
defunciones_file = "Defunciones.csv"
sospechosos_file = "Sospechosos.csv"
negativos_file = "Negativos.csv"
```

### Transform DataFrames

```python
confirmados_data = pd.read_csv(confirmados_file)
confirmados_df = pd.DataFrame(confirmados_data)
confirmados_df = confirmados_df.dropna(how='all')
confirmados_df.insert(3, "status", ["confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados","confirmados"], True) 
c1_df = confirmados_df.drop([0])
c1_df.head()
```

```python
#Nacional confirmados
na1 = confirmados_df.loc[confirmados_df['nombre'] == "Nacional"]
na1
```

```python
sospechosos_data = pd.read_csv(sospechosos_file)
sospechosos_df = pd.DataFrame(sospechosos_data)
sospechosos_df = sospechosos_df.dropna(how='all')
sospechosos_df.insert(3, "status", ["sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos","sospechosos"], True) 
s1_df = sospechosos_df.drop([0])
s1_df.head()
```

```python
na2 = sospechosos_df.loc[sospechosos_df['nombre'] == "Nacional"]
na2
```

```python
negativos_data = pd.read_csv(negativos_file)
negativos_df = pd.DataFrame(negativos_data)
negativos_df = negativos_df.dropna(how='all')
negativos_df.insert(3, "status",["negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos","negativos"], True)
n1_df = negativos_df.drop([0])
n1_df.head()
```

```python
na3 = negativos_df.loc[negativos_df['nombre'] == "Nacional"]
na3
```

```python
defunciones_data = pd.read_csv(defunciones_file)
defunciones_df = pd.DataFrame(defunciones_data)
defunciones_df = defunciones_df.dropna(how='all')
defunciones_df.insert(3, "status",["defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones","defunciones"], True)
d1_df = defunciones_df.drop([0])
d1_df.head()
```

```python
na4 = defunciones_df.loc[defunciones_df['nombre'] == "Nacional"]
na4
```

### Merge all DataFrames

```python
#Merge by state
estados_df = pd.concat([c1_df,s1_df,n1_df,d1_df], ignore_index=True)
estados_df.head()
```

```python
#Merge by nacional
nacional_df = pd.concat([na1,na2,na3,na4], ignore_index=True)
nacional_df.head()
```

### Create database connection

```python
rds_connection_string = "postgres:postgres@localhost:5432/Project2"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

### Load DataFrames into database

```python
n1_df.to_sql(name='negativos', con=engine, if_exists='append', index=False)
```

```python
s1_df.to_sql(name='sospechosos', con=engine, if_exists='append', index=False)
```

```python
d1_df.to_sql(name='defunciones', con=engine, if_exists='append', index=False)
```

```python
c1_df.to_sql(name='confirmados', con=engine, if_exists='append', index=False)
```

```python
estados_df.to_sql(name='estados', con=engine, if_exists='append', index=False)
```

```python
nacional_df.to_sql(name='nacional', con=engine, if_exists='append', index=False)
```
